```python
import csv
```

```python
csv_list = ['SIGLA,UF,NAO_RECEBE,RECEBE,SEM_RESPOSTA,TOTAL\n']
# read txt file
with open('Data/Pessoas/UF_CECAD.txt', 'r') as f:
    data = f.readlines()
    for i, line in enumerate(data):
        if '-' in line:
            csv_list.append(line.replace('-',' ').replace(' ',','))

```

### LEMBRAR DE INSERIR DATA ATUAL DA COLETA

```python
with open('Data/Pessoas/UF_CECAD.csv', 'w') as f:
    for line in csv_list:
        f.write(line)
```
